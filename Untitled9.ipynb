```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
```

```python
yash=pd.read_csv('sharma-kohli.csv')
yash
```

```python
plt.plot(yash['index'],yash['V Kohli'],color='Red',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(yash['index'],yash['RG Sharma'],color='Blue',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')
 
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
 
plt.legend(loc='upper right')
```
